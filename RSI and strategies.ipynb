```python
import nsepy as nse
import ta
import pandas as pd
from datetime import date
import matplotlib.pyplot as plt
```

```python
st=date(2015,1,1)
et=date(2019,1,1)
p=nse.get_history('INFY',st,et)
o=p['Close']
o
```

```python
import matplotlib.pyplot as plt
plt.plot(ta.rsi(o))
plt.figure(1,figsize=(90,90))
plt.axhline(y=70,color='g')
plt.axhline(y=30,color='r')
plt.show()
```
