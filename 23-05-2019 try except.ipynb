```python
import sys

lst = ['b',0,2]
for entry in lst:
    try:
        print('this entry is', entry)
        r = 1/int(entry)
    except:
        print('oops!', sys.exc_info()[0],'occured.')
        print('next entry.')
        print('***********')
print('The reciprocal of', entry, "is", r)
```
