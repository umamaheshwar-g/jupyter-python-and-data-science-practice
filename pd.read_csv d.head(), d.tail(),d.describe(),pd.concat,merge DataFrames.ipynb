GROUP-BY

```python
dc=pd.read_csv('weather_data_cities.csv', index_col='day')
dc
```

```python
g=dc.groupby('city')
g.get_group('mumbai')
c=list(set(list(dc['city'])))
for i in c:
    print(i)
    print(g.get_group(i))
```

```python
k=dc.groupby('event')
e=list(set(list(dc['event'])))
for i in e:
    print(i)
    print(k.get_group(i))
```

```python
india_weather  = pd.DataFrame({'city':['mumbai','delhi','banglore'],
                               'temperature':[32,35,30],
                               'humidity':[80,60,78]})
india_weather
```

```python
us_weather  = pd.DataFrame({'city':['newyork','chicago','orlando'],
                               'temperature':[21,14,35],
                               'humidity':[68,65,75]})
us_weather
```

```python
df=pd.concat([india_weather,us_weather],axis=0)
df
```

```python
df=pd.concat([india_weather,us_weather],axis=1)
df
```

```python
df=pd.concat([india_weather,us_weather],ignore_index=True)
df
```

```python
temperature_df = pd.DataFrame({
    'city':['mumbai','delhi','banglore','hyderabad'],
    'temperature':[32,45,30,40]})
temperature_df
```

```python
humidity_df = pd.DataFrame({
    'city':['delhi','mumbai','banglore'],
    'humidity':[58,68,70]})
humidity_df
```

```python
df = pd.merge(temperature_df, humidity_df,on='city', how='outer')
df
```

```python
df = pd.merge(temperature_df, humidity_df,on='city', how='inner')
df
```
