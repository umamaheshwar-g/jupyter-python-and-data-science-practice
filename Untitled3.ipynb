```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
haber=pd.read_csv('haberman.csv')
haber.columns=['dob','age','nodes','status']
haber
```

```python
sns.violinplot(x='age',y='status',data=haber)

plt.show()
```

```python
sns.set_style('whitegrid')
sns.pairplot(haber,hue='dob',size=3)
plt.show()
```

```python
sns.jointplot(x='status',y='age',data=haber,kind='kde')
plt.show()
sns.jointplot(x='dob',y='age',data=haber,kind='kde')
plt.show()
sns.jointplot(x='dob',y='status',data=haber,kind='kde')
plt.show()
```

```python
sns.boxplot(x='age',y='status',data=haber,fliersize=7)
plt.show()
```

```python
print("Means:")
print(np.mean(haber["dob"]))
#Mean with an outliner.
#print(np.mean(np.append(haber["30"],50)))

print(np.mean(haber["age"]))
print(np.mean(haber["status"]))

print("\nstd-dev:");
print(np.std(haber["age"]))
print(np.std(haber["dob"]))
print(np.std(haber["status"]))
```

```python
print("\nMedians:")
print(np.median(haber["30"]))

print(np.median(np.append(haber["30"],50)))
print(np.median(haber["64"]))
print(np.median(haber["1.1"]))

print("\nQualities:")
print(np.percentile(haber["30"],np.arange(0,100,25)))
print(np.percentile(haber["64"],np.arange(0,100,25)))
print(np.percentile(haber["1.1"],np.arange(0,100,25)))

print("\n90th percentiles:")
print(np.percentile(haber["30"],90))
print(np.percentile(haber["64"],90))
print(np.percentile(haber["petal.length"],90))

from statsmodels import robust
print("\nMedian Absolute Deviation:")
print(robust.mad(haber["petal.length"]))#median absolute deviation
print(robust.mad(haber["petal.length"]))
print(robust.mad(haber["petal.length"]))
```
